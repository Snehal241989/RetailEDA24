```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = '/content/sample_data/Grip_EDA.xlsx'
data = pd.read_excel(file_path)

# Set up the visualization
plt.figure(figsize=(8, 4))

# Bar chart for Profit by Region
sns.barplot(data=data, x='Region', y='Profit', estimator=sum, ci=None)
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/content/sample_data/Grip_EDA.xlsx'
data = pd.read_excel(file_path)

# Calculate the total profit for each category
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()

# Plot the profit by category
plt.figure(figsize=(10, 6))
plt.bar(profit_by_category['Category'], profit_by_category['Profit'], color=['blue', 'green', 'red'])
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.title('Total Profit by Category')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/content/sample_data/Grip_EDA.xlsx'
data = pd.read_excel(file_path)

# Calculate the total sales for each category
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()

# Plot the sales by category
plt.figure(figsize=(8, 6))
plt.bar(sales_by_category['Category'], sales_by_category['Sales'], color=['brown', 'yellow', 'blue'])
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Category')
plt.show()
```
